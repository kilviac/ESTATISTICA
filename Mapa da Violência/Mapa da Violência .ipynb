* EVOLUÇÃO DO HOMICÍDIO DE MULHERES NEGRAS NOS ESTADOS NORDESTINOS (2003-2013)

```python
df9 = pd.read_csv('hmulheresn.csv',encoding="latin2", delimiter = ',', index_col = 0) 
df9
```

```python
df9.plot.bar(figsize=(25, 5), title = 'Evolução dos homicídios das MULHERES NEGRAS nos estados nordestino', grid=True)
```

Análise: Tomando como referência os estados da Bahia e Pernambuco, que possuem os maiores índices, é possível perceber que mulheres negras morrem mais que mulheres brancas. A Bahia, ao longo dos anos de 2003 a 2013, obteve um total de 3154 homicídios de muheres, envolvendo diversos meios, desse total, 328 eram de mulheres brancas, ou seja, um pouco mais de 11%, enquanto a quantidade de mulheres negras mortas somava um total de 2.826, ou seja 89% do total de casos. A mesma coisa não difere em Pernambuco, das 3.012 mortes, 397 foram de mulheres brancas (13%), e os outros 2615 de mulheres negras (87%). Essa tendência também é encontrada na análise dos outros estados do nordeste, isso ocorre porque a soceidade brasileira é fincada no racismo patriarcal, que é um fator determinante para o número elevado de feminicídio no Brasil. O desenvolvimento econômico baseado na escravidão e exploração da população negra, aumenta o risco da violência fatal e geram condições de vida desiguais. Os indicadores sociais brasileiros, quando avaliam as perspectiva das mulheres negras, revela todo um contexto de desigualdade e risco a vida. Devido a falta de políticas que incluissem o negro na sociedade após abolição da escravidão, os negros foram historicamente empurrados para as áreas mais afastadas e com condições precárias, com isso, as mulheres negras são maioria também entre pessoas que vivem em situações precárias.

Análise da queda de homicídios das mulheres negras em PERNAMBUCO: Segundo o site "isso é feminicídio", a redução dessa taxa se deu principalmente, porque o Governador de Pernmabuco aderiu ao protocolo latino-americano de investigação de mortes violentas de mulheres, e também ao grande volume de campanhas em tv aberta, para que mulheres que sofrem agressões denunciem os agressores.

* Evolução de homicídios utilizando ARMA DE FOGO nas Capitais dos Estados Nordestinos (2004-2014)

```python
df10 = pd.read_csv('afcapitais.csv',encoding="latin2", delimiter = ';', index_col = 0) 
df10
```

```python
df10.plot.bar(figsize=(25, 5), title = 'Evolução dos homicídios utilizando arma de fogo nas capitais', grid=True)
plt.ylabel('Valores a cada mil')
```

Análise: O estado da Bahia apresentou um total de 38.836 homicídios utilizando arma de fogo dos anos de 2004 a 2014, possuindo o maior índice dentre os estados nordestinos. Sua capital, Salvador, representa aproximadamente 20% da população total, no entanto, representa 35% (13.663) da evolução dos homicídios que ocorreu no estado. Já Pernambuco, possui como capital Recife, que concentra aproximandamente 43% da população total do estado, apresentou cerca de 29% (9.713) do total de homicídios na mesma faixa de tempo, ou seja, nesse caso, não é uma regra que as capitais vão concentrar o maior índice de homicídios utilizando determinados meios.

Observando o gáfico das capitais, Fortaleza apresenta um crescimento de aproximadamente 200%, comparando os anos de 2004 a 2014. Segundo o site Agência Brasil, isso ocorreu devido a expansão do tráfico de drogas, que passou a deter territórios maiores, a ter acesso a armas, o crescimento urbano exagerado e o desenvolvimento enconômuco que provoca desigualdade.
